```python
import pandas as pd
from nsepy import get_history
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
start=datetime(2020,10,1)
end=datetime(2021,10,8)
```

```python
df=get_history(symbol='YESBANK',start=start,end=end)  
df
```

```python
df = df.reset_index()
```

```python
df1 = df[['Date', 'Turnover']]
```

```python
df1.set_index('Date', inplace=True)
```

```python
df1.head()
```

```python
from statsmodels.tsa.stattools import adfuller
#Ho: It is non stationary
#H1: It is stationary

def adfuller_test(Turnover):
    result=adfuller(Turnover)
    labels = ['ADF Test Statistic','p-value','#Lags Used','Number of Observations Used']
    for value,label in zip(result,labels):
        print(label+' : '+str(value) )
    if result[1] <= 0.05:
        print("strong evidence against the null hypothesis(Ho), reject the null hypothesis. Data has no unit root and is stationary")
    else:
        print("weak evidence against null hypothesis, time series has a unit root, indicating it is non-stationary ")
```

```python
adfuller_test(df1['Turnover'])
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
series = df1['Turnover']
decom = seasonal_decompose(series, model = 'additive',freq=12, extrapolate_trend = 12)
```

```python
trend = decom.trend
seasonal = decom.seasonal
residual = decom.resid
```

```python
plt.figure(figsize=(15,6))
plt.subplot(411)
plt.plot(df1['Turnover'], label = 'Orginal')
plt.legend(loc = 'best')
plt.subplot(412)
plt.plot(trend, label = 'Trend')
plt.legend(loc = 'best')
plt.subplot(413)
plt.plot(seasonal, label = 'Seasonal')
plt.legend(loc = 'best')
plt.subplot(414)
plt.plot(residual, label = 'Residual')
plt.legend(loc = 'best')
```

```python
df1_train = df1.iloc[:200,:]
df1_test = df1.iloc[200:, :]
```

### ARMA Model

```python
from statsmodels.tsa.arima_model import ARMA
model = ARMA(df1_train,order = (5,0))
fitted = model.fit()
fitted.summary()
```

```python
fc,se,conf =  fitted.forecast(54, alpha = 0.05)
```

```python
fc_series = pd.Series(fc, index=df1_test.index)
lower_series = pd.Series(conf[:, 0], index=df1_test.index)
upper_series = pd.Series(conf[:, 1], index=df1_test.index)
```

```python
plt.figure(figsize=(15,6))
plt.plot(df1_train, label = 'Training')
plt.plot(df1_test, label = 'Test', color = 'red')
plt.plot(fc_series, label = 'Forecasting', color = 'green')
plt.fill_between(lower_series.index, lower_series, upper_series,color='g', alpha=.05)
plt.legend(loc = 'best')
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARMA
model = ARMA(df1_train,order = (10,0))
fitted = model.fit()
fitted.summary()
```

```python
fc,se,conf =  fitted.forecast(54, alpha = 0.05)
fc_series_1 = pd.Series(fc, index=df1_test.index)
lower_series_1 = pd.Series(conf[:, 0], index=df1_test.index)
upper_series_1= pd.Series(conf[:, 1], index=df1_test.index)
plt.figure(figsize=(15,6))
plt.plot(df1_train, label = 'Training')
plt.plot(df1_test, label = 'Test', color = 'red')
plt.plot(fc_series_1, label = 'Forecasting', color = 'green')
plt.fill_between(lower_series_1.index, lower_series_1, upper_series_1,color='g', alpha=.05)
plt.legend(loc = 'best')
plt.show()
```

### ARIMA Model

```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(df1['Turnover'], order = (1,1,1))
model_fit = model.fit()
```

```python
model_fit.summary()
```

```python
df1['forecast'] = model_fit.predict(start = 200, end = 253, dynamic = True)
```

```python
df1[['Turnover','forecast']].plot(figsize = (12,8))
```

### SARIMA Model

```python
import statsmodels.api as sm
```

```python
model_1=sm.tsa.statespace.SARIMAX(df1['Turnover'],order=(1, 1, 1),seasonal_order=(1,1,1,12))
results=model_1.fit()
```

```python
df1['forecast_1'] = results.predict(start = 200, end = 253, dynamic = True)
df1[['Turnover','forecast_1']].plot(figsize = (20,8))
```
